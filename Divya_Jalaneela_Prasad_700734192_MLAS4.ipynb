2. Apply K means clustering in the dataset provided:
• Remove any null values by the mean.
• Use the elbow method to find a good number of clusters with the K-Means algorithm
• Calculate the silhouette score for the above clustering

```python
input_df2=pd.read_csv("datasets/K-Mean_Dataset.csv")
input_df2.head()
```

```python

df = input_df2.iloc[:,1:].values

new_df= SimpleImputer(missing_values=np.nan, strategy='mean')
new_df = new_df.fit(df)
df = new_df.transform(df)
```

```python
input_df2.columns
```

```python

wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('the elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('Wcss')
plt.show()
```

```python

from sklearn.cluster import KMeans
nclusters = 4 # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(df)

y_kmeans = km.predict(df)
```

```python

from sklearn import metrics
score = metrics.silhouette_score(df, y_kmeans)
print('Silhouette score:',score)
```
